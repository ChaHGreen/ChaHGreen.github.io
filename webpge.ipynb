## Combine the json files

```python
import json
import os
root='F:\A_llava\llava_3doi\\result_processing\\results\\batch_output_two_images_34'
ty='bbox' #'bbox' 'keypoint'
file_name=os.path.basename(root)
output_file=f"{root}\\{file_name}_output.json"
# f"{root}\\batch_output_{ty}_34_output.json"

file_paths = [f'{root}\movable\movable_output.json', 
              f'{root}\\rigid\\rigid_output.json', 
              f'{root}\kinematic\kinematic_output.json', 
              f'{root}\\interaction_type\\interaction_type_output.json']

# 读取文件内容
file_contents = [json.load(open(file_path, 'r')) for file_path in file_paths]

indi=ty#'bbox'   #or keypoint
# 创建新的字典
new_dict = {}

# 默认的键值
default_values = {'movable': 'n/a', 'rigid': 'n/a', 'kinematic': 'n/a', 'interaction_type': 'n/a'}

# 遍历文件内容
for content in file_contents:
    for item in content:
        # 创建键
        key = (item['img_name'], str(item[indi]))
        
        # 如果键已经在字典中，添加新的描述和独特的键
        if key in new_dict:
            new_dict[key]['description'].append(item['description'])
            new_dict[key].update({k: v for k, v in item.items() if k not in ['img_name', indi, 'description']})
        # 如果键不在字典中，创建新的条目
        else:
            new_dict[key] = {
                'img_name': item['img_name'],
                f'{indi}': item[indi],
                'description': [item['description']],
            }
            new_dict[key].update(default_values)
            new_dict[key].update({k: v for k, v in item.items() if k not in ['img_name', indi, 'description']})

# 将字典转换为列表
new_list = list(new_dict.values())

# 将列表保存为新的JSON文件
with open(output_file, 'w') as f:
    json.dump(new_list, f,indent=2)
```

```python
from PIL import Image,ImageDraw
import math
def draw_star(image, center):
    # 创建五角星的顶点
        draw = ImageDraw.Draw(image)
        # 创建五角星的顶点
        color='red'
        size=10
        points = []
        for i in range(5):
            angle_deg = 72 * i - 90  # 五角星每个顶点相隔72度，起始点向上
            angle_rad = math.radians(angle_deg)
            x = center[0] + size * math.cos(angle_rad)
            y = center[1] + size * math.sin(angle_rad)
            points.append((x, y))

        # 五角星的内顶点
        for i in range(5):
            angle_deg = 72 * i - 90 + 36  # 内顶点偏移36度
            angle_rad = math.radians(angle_deg)
            x = center[0] + size * math.cos(angle_rad) * 0.5  # 内顶点的半径是外顶点的一半
            y = center[1] + size * math.sin(angle_rad) * 0.5
            points.insert(2 * i + 1, (x, y))

        # 绘制实心五角星
        draw.polygon(points, fill=color)
        return image
```
